```python
print(df["Order Date"].value_counts())
```

```python

def parse_time(time):
    try:
      return pd.to_datetime(time, format='%m%d/%Y %H%M%S')
    except ValueError:
        try:
            return pd.to_datetime(time, format='%m/%d/%Y %H:%M')
        except ValueError:
            return pd.NaT

df['Order Date'] = df['Order Date'].apply(parse_time)
```

```python
df["Price Each"]=df['Price Each'].astype(float)
```

```python
print(df["Price Each"].value_counts())
cleaned_df=df.loc[~(df["Price Each"]=="Price Each")]
```

```python

cleaned_df.dropna(inplace=True)
cleaned_df["Order Date"]=pd.to_datetime(cleaned_df['Order Date']
                                                    # Attempt to infer format of each date
                                                    ,infer_datetime_format=True,
                                                    # Return NA for rows where conversion failed
                                                                    errors = 'coerce')
cleaned_df["Order ID"]=cleaned_df['Order ID'].astype(int)
cleaned_df["Quantity Ordered"]=cleaned_df['Quantity Ordered'].astype(int)
cleaned_df["Price Each"]=cleaned_df['Price Each'].astype(float)
cleaned_df.info()
```

```python
all_phone_products=cleaned_df[cleaned_df["Product"].isin(["Vareebadd Phone","Google Phone","iPhone","Wired Headphones","Apple Airpods Headphones"])]
grouped_data=all_phone_products.groupby(["Purchase Address"])[["Price Each"]].sum()
grouped_data.sort_values(by="Price Each",ascending=False)
```
