# Statistics

### Min and Max

```python
a = np.array([[2,10,20],[80,43,31],[22,43,10]])  
  
print("The original array:\n")  
print(a)
print("\nThe minimum element among the array:",np.amin(a))  
print("The maximum element among the array:",np.amax(a))  

print("\nThe minimum element among the rows of array\n",np.amin(a,axis=1).reshape(1,-1).T)  
print("The maximum element among the rows of array\n",np.amax(a,axis=1).reshape(1,-1).T)  

print("\nThe minimum element among the columns of array",np.amin(a,axis=0))
print("The maximum element among the columns of array",np.amax(a,axis=0))
```

## Peak-to-Peak aka Range

```python
a = np.array([[2,10,20],[80,43,31],[22,43,10]])  
  
print("Original array:\n",a) 
b=np.ptp(a,1).reshape(1,-1)
print("\nptp value along axis 1:\n",b.T)  

print("ptp value along axis 0:",np.ptp(a,0))  
```

## Percentiles

```python
a = np.array([2,100,20,80,43,31,22,43,-60])  
Q1,Q3 = np.percentile(a, [25,75])
IQR = (Q3 - Q1)
print("Q1:",Q1)
print("Q3:",Q3)
print("IQR:",IQR)
```

## Detecting outliers

```python
a = np.array([2,100,20,80,43,31,22,43,-60])
def outliers(a):
    Q1,Q3 = np.percentile(a, [25,75])
    lower_bound=Q1-1.5*IQR
    upper_bound=Q3+1.5*IQR
    for i in range(len(a)):
        if a[i] > upper_bound or a[i] < lower_bound:
            a[i]=True
        else:
            a[i]=False
outliers(a)
print(a)
```

```python
#define our dataset
dataset= [11,9,10,11,8,11,12,13,12,11,16,10,110,120,11,14,10,11,12,13,14,11,9,10.8,7,5,6,7,10,12,11,-120,9,10,11,1,5,9,7]
```

```python
outlier=[]
def detect_outlires(data):
    threshold = 3 #3 std  deviation from mean for an anomaly to be considered as an outlier.
    mean= np.mean(data)
    std= np.std(data)

    for i in data:
        zscore = (i - mean)/std
        if np.abs(zscore) >threshold:
            outlier.append(i)
    return outlier
print(detect_outlires(dataset))
```

## Probability

```python
import pandas as pd
df=pd.read_csv('CompanyABCProfit.csv')
df.head()
```

```python
import matplotlib.pyplot as plt
plt.hist(df["Profit(Rs '000)"])
plt.show()
```

### What percent of the years did the company finish the year with profit less than 500

```python
from scipy.stats import norm
x=norm.cdf(500, df["Profit(Rs '000)"].mean(), df["Profit(Rs '000)"].std())
print(x)
```

### What percent of the years did the company finish the year with profit more than 500

```python
out=1-norm.cdf(500, df["Profit(Rs '000)"].mean(), df["Profit(Rs '000)"].std())
print(out)
```

### Between 500 and 1000
cdf(1000)-cdf(500)

```python
out1=norm.cdf(1000, df["Profit(Rs '000)"].mean(), df["Profit(Rs '000)"].std())-norm.cdf(500, df["Profit(Rs '000)"].mean(), df["Profit(Rs '000)"].std())
print(out1)
```

 ### the inverse of cumulative distribution function
 ### aka ppf

### What profit are 60% of the years less than?

```python
prof=norm.ppf(0.6,df["Profit(Rs '000)"].mean(), df["Profit(Rs '000)"].std())
print(prof)
```

## PMF

```python
# Example data (discrete distribution)
data = np.random.randint(1, 7, size=1000)  # simulate 1000 rolls of a fair six-sided die

# Compute PMF
unique, counts = np.unique(data, return_counts=True)
pmf = counts / len(data)

# Compute CDF
cdf = np.cumsum(pmf)

# Plot PMF
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.bar(unique, pmf, width=0.4, align='center', color='skyblue')
plt.xlabel('Outcome')
plt.ylabel('Probability')
plt.title('Probability Mass Function (PMF)')

# Plot CDF
plt.subplot(1, 2, 2)
plt.step(unique, cdf, where='mid', color='orange')
plt.xlabel('Outcome')
plt.ylabel('Cumulative Probability')
plt.title('Cumulative Distribution Function (CDF)')

plt.tight_layout()
plt.show()
```
