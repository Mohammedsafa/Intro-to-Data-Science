## Import The needed library 

```python
import pandas as pd 
import numpy as np 
from scipy.stats import spearmanr
from scipy import  stats
import matplotlib.pyplot as plt
from scipy.stats import binom
from scipy.stats import poisson
from scipy.stats import uniform
from scipy.stats import expon
from scipy.stats import chi2
from scipy.stats import t
from scipy.stats import f
from scipy.stats import lognorm
```

### Initialize random array

```python
np.random.seed(41)
x= np.random.randint(0,1000,size=20)
y= np.random.randint(0,1000,size=20)
print ("x array:", x)
print("#"*50)
print("y array:", y)
```

### Apply co-variance 

```python
covariance = np.cov(x, y)[0][1]
print("Covariance:", covariance)
```

### Apply Pearson correlation coefficient

```python
correlation_coefficient = np.corrcoef(x, y)[0][1]
print("Pearson correlation coefficient:", correlation_coefficient)
```

### Apply Spearman’s Rank Correlation

```python
rank_X = [rank + 1 for rank in np.argsort(x)]
rank_Y = [rank + 1 for rank in np.argsort(y)]

# Calculate Spearman's correlation coefficient
spearman_coefficient, _ = spearmanr(rank_X, rank_Y)
print("Spearman's rank correlation coefficient:", spearman_coefficient)
```

## Kinds Of Distribution 

### 1-Normal Distribution (Gaussion)

- Description: The normal distribution is a symmetric, bell-shaped distribution characterized by its mean and standard deviation. It is widely used in statistical inference and modeling due to its simplicity and because many natural phenomena follow this distribution.
- Pros: It is mathematically tractable, and many statistical methods assume normality. It is also the foundation for many other statistical distributions and techniques.


```python
mu = 0
sigma = 1
num_samples = 1000

data = np.random.normal(mu, sigma, num_samples)

plt.hist(data, bins=30, density=True, alpha=0.6, color='g')
plt.title('Normal Distribution')
plt.show()
```

### 2-Binomial Distribution

- Description: The binomial distribution describes the probability of a certain number of successes in a fixed number of independent Bernoulli trials.
- Pros: It's simple and easy to understand. It's widely applicable in various fields, including biology, sociology, and quality control.

```python
n = 20  # number of trials
p = 0.5  # probability of success

x = np.arange(0, n+1)
binomial_pmf = binom.pmf(x, n, p)

plt.bar(x, binomial_pmf)
plt.title('Binomial Distribution')
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.show()
```

### 3-Poisson Distribution:

- Description: The Poisson distribution models the number of events occurring in a fixed interval of time or space, given a constant average rate of occurrence and independence of events.
- Pros: It's useful for modeling rare events and counting processes, such as the number of arrivals at a service point.

```python
mu = 3  # average rate of occurrence

x = np.arange(0, 10)
poisson_pmf = poisson.pmf(x, mu)

plt.bar(x, poisson_pmf)
plt.title('Poisson Distribution')
plt.xlabel('Number of Events')
plt.ylabel('Probability')
plt.show()
```

### 4-Uniform Distribution

- Description: The uniform distribution has constant probability.
- Pros: It's simple and often used for modeling situations where all outcomes are equally likely.

```python
low = 0
high = 10

x = np.linspace(low, high, 100)
uniform_pdf = uniform.pdf(x, low, high-low)

plt.plot(x, uniform_pdf)
plt.title('Uniform Distribution')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.show()
```

### 5-Exponential Distribution

- Description: The exponential distribution models the time until an event occurs, given a constant rate of occurrence.
- Pros: It's commonly used for modeling waiting times and survival analysis.

```python
scale = 1  # scale parameter

x = np.linspace(0, 5, 100)
exponential_pdf = expon.pdf(x, scale=scale)

plt.plot(x, exponential_pdf)
plt.title('Exponential Distribution')
plt.xlabel('Time')
plt.ylabel('Probability Density')
plt.show()
```

### 6-Chi-Square Distribution

- Description: The chi-square distribution arises in tests of statistical significance and in the construction of confidence intervals.
- Pros: It's used in hypothesis testing and goodness-of-fit tests.

```python
df = 3  # degrees of freedom

x = np.linspace(0, 20, 100)
chi2_pdf = chi2.pdf(x, df)

plt.plot(x, chi2_pdf)
plt.title('Chi-Square Distribution')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.show()
```

### 7-Student's t-Distribution 

- Description: The Student’s t-distribution is a continuous probability distribution that describes the relationship between observed and expected, is used in hypothesis testing and constructing confidence intervals for the mean of a normally distributed population when the sample size is small and the population standard deviation is unknown.
- Pros: It's robust against deviations from normality and is widely used in statistical inference.

```python
df = 10  # degrees of freedom

x = np.linspace(-5, 5, 100)
t_pdf = t.pdf(x, df)

plt.plot(x, t_pdf)
plt.title("Student's t-Distribution")
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.show()
```

### 8-F-Distribution 

- Description: The F-distribution arises in the analysis of variance and in the construction of confidence intervals for variance components.
- Pros: It's used in analysis of variance (ANOVA) and regression analysis.

- - ANOVA, or Analysis of Variance, is a statistical method used to compare the means of three or more groups to determine if there are statistically significant differences between them. It extends the t-test, which is used to compare the means of two groups.

```python
dfn = 5  # numerator degrees of freedom
dfd = 10  # denominator degrees of freedom

x = np.linspace(0, 5, 100)
f_pdf = f.pdf(x, dfn, dfd)

plt.plot(x, f_pdf)
plt.title('F-Distribution')
plt.xlabel('Value')
plt.ylabel
plt.show()
```

### 9-Lognormal Distribution

- Description: The lognormal distribution describes a random variable whose logarithm is normally distributed.
- Pros: It's often used to model the distribution of wealth, incomes, and stock prices.

```python
s = 0.8  # shape parameter

x = np.linspace(0, 5, 100)
lognormal_pdf = lognorm.pdf(x, s)

plt.plot(x, lognormal_pdf)
plt.title('Lognormal Distribution')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.show()
```

### Implement Z-test 

```python
population_mean = 150
population_std = 10
sample_mean= 155
sample_size = 50
z_score = (sample_mean - population_mean)/ ( population_std / np.sqrt(sample_size))
p_value = 2*(1 - stats.norm.cdf(np.abs(z_score)))
print("Z-score:", z_score)
print("P-value:", p_value)
```

```python
from statsmodels.stats.weightstats import ztest
np.random.seed(14)

pop_age = np.random.randint(15, 80, size=50)
pop_mean = np.mean(pop_age)

sample_age_size = 10
sample_age = np.random.choice(pop_age, size=sample_age_size)

z_score, p_value = ztest(x1=sample_age, value=pop_mean)

print("Z-score:", z_score)
print("P-value:", p_value)
```

### Implement T-Test

```python
t_statistic, p_value = stats.ttest_1samp(sample_age, pop_mean)
print("T-statistic:", t_statistic)
print("P-value:", p_value)
```

### Implement CHI-Square test

```python
observed_values = np.array([[30, 10, 20], [20, 15, 25]])

row_totals = np.sum(observed_values, axis=1)
column_totals = np.sum(observed_values, axis=0)
grand_total = np.sum(observed_values)

expected_values = np.outer(row_totals, column_totals) / grand_total

chi2_statistic = np.sum((observed_values - expected_values)**2 / expected_values)

print("Chi-square statistic:", chi2_statistic)
```

```python
observed_values = np.array([[30, 10, 20], [20, 15, 25]])

chi2_statistic, p_value, degrees_of_freedom, expected_values = stats.chi2_contingency(observed_values)

print("Chi-square statistic:", chi2_statistic)
print("P-value:", p_value)
print("Degrees of freedom:", degrees_of_freedom)
print("Expected values:")
print(expected_values)
```
